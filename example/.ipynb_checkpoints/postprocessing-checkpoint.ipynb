```python
#Get data to plot the BSE absorption spectra
opgd0,name,ex_en,ex_os=get_opg(db, mid)
bfreq,beps1,beps2,name=get_bse(db, mid)
print('Optical gap:','%5.2f' %opgd0)
```

```python
##Retrieve QP energies and INCAR paramters for converged GW calculation  
qpe, incar = get_qpe(db, {'material_id': mid, "task_label": { "$regex": 'scGW'}, 'ifconv': True})
igap, dgap = get_gap_from_dict(qpe)
```

```python
#Exciton Binding Energy 
ebe=dgap-opgd0
print('Exciton Binding Energy:', '%5.2f' %ebe)
```

```python
#Get data to plot the RPA absorption spectra
rfreq,reps1,reps2,name=get_rpa(db, mid)
```

```python
#Dielectric Function for different polarization axis
#BSE absorption spectra
beps2x=column(beps2, 0)
beps2y=column(beps2, 1)
beps2z=column(beps2, 2)
#RPA absorption spectra
reps2x=column(reps2, 0)
reps2y=column(reps2, 1)
reps2z=column(reps2, 2)
```

```python
#Plot BSE absorption spectra
plt.xlim(opgd0-1.0,opgd0+2)
plt.plot(bfreq,beps2x,c='r',ls='-', lw=2,label=r'$\epsilon_2^{xx}$')
plt.plot(bfreq,beps2y,c='b',ls='--', lw=2,label=r'$\epsilon_2^{yy}$')
plt.plot(bfreq,beps2z,c='g',ls='-.', lw=2,label=r'$\epsilon_2^{zz}$')
plt.xlabel('frequency in eV', fontsize=15)
plt.ylabel(r'$\epsilon_2$',fontsize=15)
plt.legend()
plt.show()
```

```python
#Plot RPA absorption spectra
plt.xlim(opgd0-1.0,opgd0+2)
plt.plot(rfreq,reps2x,c='r',ls='-', lw=2,label=r'$\epsilon_2^{xx}$')
plt.plot(rfreq,reps2y,c='b',ls='--', lw=2,label=r'$\epsilon_2^{yy}$')
plt.plot(rfreq,reps2z,c='g',ls='-.', lw=2,label=r'$\epsilon_2^{zz}$')
plt.xlabel('frequency in eV', fontsize=15)
plt.ylabel(r'$\epsilon_2$',fontsize=15)
plt.legend()
plt.show()
```
